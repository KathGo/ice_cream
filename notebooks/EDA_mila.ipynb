```python
from datetime import datetime
date =[]
date_day =[]
date_month=[]
date_year=[]
for i in range(0, len(raw_df[' yyyyddd'])):
   day_year=str(raw_df[' yyyyddd'][i])
   day_num = day_year[-3:] 
   year_num = day_year[:4] 
   #print(year_num) 
   #print("The day number : " + str(day_num))
  
   # adjusting day num
   day_num.rjust(3 + len(day_num), '0')
   
   # converting to date
   res = datetime.strptime(year_num + "-" + day_num, "%Y-%j").strftime("%m-%d-%Y")
   date.append(res)
   str_day=str(res)
   #print(res)
   date_day.append(res[3:5])
   date_year.append(res[6:])
   date_month.append(res[0:2])
raw_df = raw_df.assign(
    date = date,
    day = date_day,
    month = date_month,
    year = date_year
    
    )
raw_df.head(10)
```

```python
raw_df=raw_df.drop([' yyyyddd'], axis=1)
raw_df.head(10)
```

```python
raw_df= raw_df.rename(columns={' (0) Northern_Hemisphere': 'Northern_Hemisphere', ' (1) Beaufort_Sea' : 'Beaufort_Sea' , ' (2) Chukchi_Sea' :' Chukchi_Sea',
       ' (3) East_Siberian_Sea' :'East_Siberian_Sea', ' (4) Laptev_Sea' : 'Laptev_Sea', ' (5) Kara_Sea' : 'Kara_Sea', ' (6) Barents_Sea': 'Barents_Sea',
       ' (7) Greenland_Sea': 'Greenland_Sea ',' (8) Baffin_Bay_Gulf_of_St._Lawrence': 'Baffin_Bay_Gulf_of_St._Lawrence', ' (9) Canadian_Archipelago': 'Canadian_Archipelago', ' (10) Hudson_Bay': 'Hudson_Bay',
       '  (11) Central_Arctic':'Central_Arctic', 
       '(12) Bering_Sea': 'Bering_Sea', ' (13) Baltic_Sea':'Baltic_Sea', ' (14) Sea_of_Okhotsk' :'Sea_of_Okhotsk',
       ' (15) Yellow_Sea': 'Yellow_Sea', ' (16) Cook_Inlet':'Cook_Inlet'})
raw_df.columns
```
